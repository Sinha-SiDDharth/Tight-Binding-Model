We have epsilon = onsite energy of the lattice points. And then Alternating Overlap Integrals Ess and Ess' assume t and t' in our case.

So we can write three Hamiltonians. One for the On-site. Another one for the Left connection and another for the Right connection. Then we do use of the Bloch theorem to get the eigenvalues of h(k) matrix...getting our dispersion relation and hence the bandstructure... 




```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(["science", "notebook" ,"grid"])

# Parameters
t = -1.0       # hopping within unit cell (A to B)
t_prime = -2.0 # hopping between unit cells (B to A)
epsilon = 0.0  # on-site energy
a = 1.0        # lattice constant

# k values in First Brillouin Zone: [-/a, /a]
num_k = 100
k_vals = np.linspace(-np.pi/a, np.pi/a, num_k)

# Band energies
E_plus = []
E_minus = []

for k in k_vals:
    # Construct the 2x2 Bloch Hamiltonian h(k)
    Hk = np.array([
        [epsilon, t + t_prime * np.exp(-1j * k * a)],
        [t + t_prime * np.exp(1j * k * a), epsilon]
    ])
    
    # Get eigenvalues (real, since Hermitian)
    eigenvalues = np.linalg.eigvalsh(Hk)

    # Store both bands
    E_minus.append(eigenvalues[0])
    E_plus.append(eigenvalues[1])

# Plot the band structure
plt.plot(k_vals/np.pi, E_plus, label='Upper Band (+)', color='red')
plt.plot(k_vals/np.pi, E_minus, label='Lower Band (-)', color='blue')
plt.xlabel(r"$\frac{ka}{\pi}$")
plt.ylabel("Energy")
plt.title("Numerical Band Structure of 1D Dimerized Chain (Peierls Distortion)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook' , "grid"])

# Parameters
a = 2
b = 1

# k in units of /a
k_vals = np.linspace(-1, 1, 21)

# Energy dispersion relation
E = np.sqrt(a**2 + b**2 + 2*a*b*np.cos(np.pi * k_vals))

# Plot both +E and -E branches
plt.plot(k_vals, E, 'b', linewidth=2.0, label='+E(k)' ,color='red')
plt.plot(k_vals, -E, 'b', linewidth=2.0, label='-E(k)')

plt.xlabel(r'$k$ (in units of $\pi/a$)', fontsize=14)
plt.ylabel('Energy (eV)', fontsize=14)
plt.title("Analytical Band Structure of Dimerized Chain")
plt.legend()
plt.show()
```
